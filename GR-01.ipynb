*d) The locus of events whose interval ∆$s^2$ from the origin is -1$m^2$*

For this 2D diagram, we have $∆s^2=-(∆t)^2+(∆x)^2$ or $(∆t)^2=(∆x)^2+1$

We can consider this equivalent to $y^2=x^2+1$ for the purposes of plotting.At this point we zoom out to see the behavior better.

```python
import matplotlib.pyplot as plt
import numpy as np
t=np.arange(-5.0,5.0,0.2) # The "ticker", adjust for precision. 

plt.figure(figsize=(10,10)) #make it larger
plt.gca().set_aspect('equal') #make it square so the axes are easier to see.

plt.xlabel("x (m)")
plt.ylabel("t (m)")

plt.axis([-4,4,-4,4])

plt.axvline(x=1)
plt.plot(t,10*t-5,'-g')
plt.plot(t*np.cos(0.456),t*np.sin(0.456),'-k')
plt.plot(t*np.sin(0.456),t*np.cos(0.456),'-k')
plt.plot(t,(t**2+1)**(1/2),'--m')
plt.plot(t,-(t**2+1)**(1/2),'--m') #the square root has both + and - solutions. 
plt.axvline(x=0,color='black')
plt.axhline(y=0,color='black') #add central axes

plt.show()
```

*d) Compute the interval ∆$s^2$ between the events at which the detectors emitted their signals, using both the coordinates of O and $\tilde{O}$*

For the altered reference frame, not so much. The coordinates of release were found to be about (-6,6.2) and (0.2,1.5) which are estimates, admittedly. First one gives -2.44 and the other gives -2.21. Clearly we need to use more precise mathematics, as these should be the same, but they are close enough to justify at the moment. 

What we need are an exact coordinate transform, which we shall attain LATER. Or right now. 

This is the Lorentz transformation. 

In 2D (x and t only) this is written as t'=α(t-vx) and x'=α(x-vt). v in this case is the velocity of the reference frame, and α=1/√(1-$v^2$). Since v = 0.75, we have α=1.512.
Now, the two points we have are (2,2.5) and (-2,2.5), so x=±2 and t=2.5. 
This gives us for (2,2.5) -> (0.189,1.512)
And for (-2,2.5)-> (-5.858,6.047) 
Both of which have ∆$s^2$ of -2.25, as they should.
